```python
from ShakerMakerResults import ShakerMakerData

import numpy as np
import os

import h5py
import matplotlib.pyplot as plt
```

## Surface simple

```python
folder= '/home/pxpalacios/Dropbox/resources/Gaussian_Surface'

h5drm = os.path.join(folder, 'Gaussian_Surface.h5drm')
h5gf  = os.path.join(folder, 'Gaussian_Surface_gf.h5')
h5map  = os.path.join(folder, 'Gaussian_Surface_map.h5')


surface_gaussian = ShakerMakerData(h5drm)
surface_gaussian.load_gf_database(h5gf)
surface_gaussian.load_map(h5map)
```

```python
# resample 
surface_gaussian_resample = surface_gaussian.resample(dt=0.01)
surface_gaussian_resample.plot_node_response(
                                node_id='QA', 
                                data_type='vel', 
                                figsize=(8, 6),
                                factor=1.0,
                                )
```

```python
# get_window

surface_gaussian_window= surface_gaussian.get_window(t_start=0.0, t_end=10.0)

surface_gaussian_window.plot_node_response(
                                target_pos=[6 , 8,  0.00],
                                data_type='vel', 
                                figsize=(8, 6),
                                factor=1.0,
                                )
```

```python
surface_gaussian_window.plot_domain( 
    xyz_origin=None,
    label_nodes='corners',
    show_calculated=False,
    figsize=(8, 6),
    axis_equal=True,
)
```

```python
surface_gaussian_window.plot_domain_calculated_t0(
    subfault='all',
    xyz_origin=None,
    show_calculated_only=True,
    figsize=(8, 6),
    axis_equal=True,
    cmap="viridis",
)
```

```python
surface_gaussian_window.plot_gf_connections(
                        node_id='QA',
                        xyz_origin=None,
                        label_nodes='donor_receivers',
                        figsize=(8,6),
                        axis_equal=True)  
```

```python
# Por ID
surface_gaussian_window.plot_node_response( 
                        node_id=['QA' , 0,100], 
                        data_type='vel', 
                        figsize=(8, 6),
                        factor=1.0,
                        )
```

```python
# Por ID
surface_gaussian_window.plot_node_newmark( 
                        node_id=['QA' , 0,100], 
                        data_type='accel', 
                        spectral_type='PSa',
                        figsize=(8, 6),
                        factor=1/9.81,
                        )
```

```python
surface_gaussian_window.plot_node_tensor_gf(
                        node_id=['QA'],
                        subfault=0, 
                        figsize=(8, 6),
                        )
```

```python
surface_gaussian_window.plot_node_arias(              
                    node_id=['QA' , 0,100],
                    data_type     = 'accel',
                    figsize=(8, 6),
                    factor=1,
                    )
```

```python
# Plot surface snapshot
surface_gaussian_window.plot_surface(  time=5.0, 
                            component='resultant', 
                            data_type='vel', 
                            cmap='RdBu_r', 
                            figsize=(8, 8),
                            elev=30, azim=-60, s=20, alpha=0.9,
                            axis_equal=True,)
    
```
